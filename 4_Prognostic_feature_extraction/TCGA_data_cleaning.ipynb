```python
tcga_tf_clinical.index = tcga_tf_clinical["ID"]
```

```python
tcga_tf_clinical["vital_status"].value_counts()
```

```python
dead = tcga_tf_clinical[tcga_tf_clinical["vital_status"] == "Dead"]
alive = tcga_tf_clinical[tcga_tf_clinical["vital_status"] == "Alive"]
```

```python
alive_2y = alive[alive["days_to_last_follow_up"] >= 365*2]
```

```python
alive_2y["days_to_last_follow_up"].isna().any()
```

```python
print(len(dead[dead["days_to_death"] < 365*2]))
print(len(dead[dead["days_to_death"] >= 365*2]))
print(len(alive_2y))
```

```python
dead_2y = dead[dead["days_to_death"] < 365*2]
alive_2y_but_dead = dead[dead["days_to_death"] >= 365*2]
alive_2y = alive[alive["days_to_last_follow_up"] >= 365*2]
```

```python
dead_2y["Two_year_survival"] = "NO"
alive_2y_but_dead["Two_year_survival"] = "Yes"
alive_2y["Two_year_survival"] = "Yes"
```

```python
df_2y = pd.concat([dead_2y, alive_2y_but_dead, alive_2y], ignore_index=True)
```
